```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mushroom.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
df = df.apply(le.fit_transform)    # label encoder = alphabet ko numbers me
```

```python
df
```

```python
data = df.values
data, data.shape
```

```python
train_data = data[:, 1:]
train_label = data[:, 0]
```

```python
split = 0.2      # matlab 20% or 80% me split kiya
train_X = train_data[:int((1 - split)*data.shape[0]), :]
test_X = train_data[int((1 - split)*data.shape[0]):, :]
train_Y = train_label[:int((1 - split)*data.shape[0])]
test_Y = train_label[int((1 - split)*data.shape[0]):]
```

```python
print(train_X.shape, train_Y.shape, test_X.shape, test_Y.shape)
```

```python
def prior_prob(y_label, label):
    return np.sum(y_label == label)/y_label.shape[0]
```

```python
prior_prob(test_Y, 1)
```

```python
def conditional_prob(X_train, Y_train, col,  value, label):
    X_train = X_train[Y_train == label]
    
    return np.sum(X_train[:, col] == value)/np.sum(Y_train == label)
    
```

```python
def posterior(X_train, Y_train, X_test):
    classes = np.unique(Y_train)    # class 0,1 ke liye chalega
    post_prob = []
    
    for label in classes:
        likelihood = 1
        for col in range(X_train.shape[1] ):
            c_prob = conditional_prob(X_train, Y_train, col, X_test[col], label)
            likelihood *= c_prob 
            
        prior = prior_prob(Y_train, label)
        
        post_prob.append(likelihood * prior)   # for label = 0 and 1
    
    return np.argmax(post_prob)
        
```

```python
print(posterior(train_X, train_Y, test_X[1]), test_Y[1])
```

```python
def accuracy(X_train, Y_train, test_X, test_Y):
    pred = []
    
    for i in test_X:
        y_pred = posterior(X_train, Y_train, i)
        pred.append(y_pred)
    
    print(f'Accuracy is: {np.sum(pred == test_Y)/test_Y.shape[0]}')
    
```

```python
accuracy(train_X, train_Y, test_X, test_Y)
```
